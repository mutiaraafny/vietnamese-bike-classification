```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import load_img, img_to_array, to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import os
import random
import numpy as np
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# --- FUNGSI UNTUK MENGIDENTIFIKASI FILE RUSAK ---
def identify_corrupted_files(directory):
    """Fungsi untuk mengidentifikasi file yang rusak tanpa menghapus"""
    corrupted_files = []
    valid_files = []
    
    print("Memindai file gambar...")
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                file_path = os.path.join(root, file)
                try:
                    with Image.open(file_path) as img:
                        img.verify()
                    with Image.open(file_path) as img:
                        img.load()
                    valid_files.append(file_path)
                except (IOError, SyntaxError, ValueError) as e:
                    print(f"File rusak ditemukan: {file_path}")
                    corrupted_files.append(file_path)
    
    print(f"File valid: {len(valid_files)}")
    print(f"File rusak: {len(corrupted_files)}")
    return corrupted_files, valid_files

def load_and_prepare_data(valid_files, image_size, max_samples_per_class=3000):
    """Load dan prepare data dengan class balancing"""
    print("Menganalisis struktur dataset...")
    
    bike_files = []
    motorbike_files = []
    
    for filepath in valid_files:
        path_parts = filepath.split('/')
        for part in path_parts:
            if part.lower() == 'bike':
                bike_files.append(filepath)
                break
            elif part.lower() == 'motorbike':
                motorbike_files.append(filepath)
                break
    
    print(f"File bike ditemukan: {len(bike_files)}")
    print(f"File motorbike ditemukan: {len(motorbike_files)}")
    
    if len(bike_files) == 0 or len(motorbike_files) == 0:
        print("Deteksi otomatis gagal, menggunakan struktur path...")
        bike_files = [f for f in valid_files if '/bike/' in f and '/motorbike/' not in f]
        motorbike_files = [f for f in valid_files if '/motorbike/' in f]
        print(f"Dengan deteksi path - Bike: {len(bike_files)}, Motorbike: {len(motorbike_files)}")
    
    if len(bike_files) == 0 or len(motorbike_files) == 0:
        raise ValueError("Gagal mendeteksi kedua kelas. Periksa struktur folder dataset.")
    
    # Batasi sampel dan balance
    if len(bike_files) > max_samples_per_class:
        bike_files = random.sample(bike_files, max_samples_per_class)
    if len(motorbike_files) > max_samples_per_class:
        motorbike_files = random.sample(motorbike_files, max_samples_per_class)
    
    min_samples = min(len(bike_files), len(motorbike_files))
    bike_files = bike_files[:min_samples]
    motorbike_files = motorbike_files[:min_samples]
    
    print(f"Dataset seimbang - masing-masing kelas: {min_samples} samples")
    
    all_files = bike_files + motorbike_files
    all_labels = [0] * len(bike_files) + [1] * len(motorbike_files)
    
    return all_files, all_labels

def load_images_in_batches(file_list, labels, image_size, batch_size=1000):
    """Load images dalam batch untuk menghemat memori"""
    images = []
    processed_labels = []
    
    print(f"Loading {len(file_list)} images dalam batches...")
    
    for i in range(0, len(file_list), batch_size):
        batch_files = file_list[i:i+batch_size]
        batch_labels = labels[i:i+batch_size]
        
        print(f"Processing batch {i//batch_size + 1}/{(len(file_list)-1)//batch_size + 1}")
        
        for file_path, label in zip(batch_files, batch_labels):
            try:
                img = load_img(file_path, target_size=image_size)
                img_array = img_to_array(img) / 255.0
                images.append(img_array)
                processed_labels.append(label)
            except Exception as e:
                print(f"Skip file rusak: {file_path} - {e}")
                continue
    
    print(f"Successfully loaded {len(images)} images")
    return np.array(images), np.array(processed_labels)

def plot_training_history(history):
    """Plot training history dengan style yang lebih baik"""
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs_range = range(len(acc))
    
    plt.figure(figsize=(15, 5))
    
    # Accuracy plot
    plt.subplot(1, 3, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy', marker='o', linewidth=2)
    plt.plot(epochs_range, val_acc, label='Validation Accuracy', marker='s', linewidth=2)
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.grid(True, alpha=0.3)
    plt.ylim([0.8, 1.0])
    
    # Loss plot
    plt.subplot(1, 3, 2)
    plt.plot(epochs_range, loss, label='Training Loss', marker='o', linewidth=2)
    plt.plot(epochs_range, val_loss, label='Validation Loss', marker='s', linewidth=2)
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid(True, alpha=0.3)
    
    # Overfitting analysis
    plt.subplot(1, 3, 3)
    overfitting_gap = np.array(acc) - np.array(val_acc)
    plt.plot(epochs_range, overfitting_gap, label='Accuracy Gap (Train - Val)', 
             marker='o', linewidth=2, color='red')
    plt.axhline(y=0, color='black', linestyle='--', alpha=0.5)
    plt.legend()
    plt.title('Overfitting Analysis')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy Difference')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def evaluate_model_comprehensive(model, val_images, val_labels, class_names):
    """Evaluasi model yang komprehensif"""
    print("\n" + "="*50)
    print("EVALUASI MODEL KOMPREHENSIF")
    print("="*50)
    
    # Prediksi
    predictions = model.predict(val_images, verbose=0)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(val_labels, axis=1)
    
    # Classification Report
    print("\nüìä CLASSIFICATION REPORT:")
    print("-" * 40)
    report = classification_report(true_classes, predicted_classes, 
                                 target_names=class_names, 
                                 output_dict=True)
    print(classification_report(true_classes, predicted_classes, 
                              target_names=class_names))
    
    # Confusion Matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    
    plt.figure(figsize=(15, 5))
    
    # Plot 1: Confusion Matrix
    plt.subplot(1, 3, 1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # Plot 2: Per-class Accuracy
    plt.subplot(1, 3, 2)
    class_accuracies = [report[class_name]['f1-score'] for class_name in class_names]
    colors = ['skyblue', 'lightcoral']
    bars = plt.bar(class_names, class_accuracies, color=colors)
    plt.title('Per-Class F1-Score')
    plt.ylabel('F1-Score')
    plt.ylim([0.95, 1.0])
    
    # Tambahkan nilai di atas bar
    for bar, acc in zip(bars, class_accuracies):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
                f'{acc:.4f}', ha='center', va='bottom')
    
    # Plot 3: Prediction Confidence Distribution
    plt.subplot(1, 3, 3)
    max_probs = np.max(predictions, axis=1)
    plt.hist(max_probs, bins=20, alpha=0.7, color='green', edgecolor='black')
    plt.title('Prediction Confidence Distribution')
    plt.xlabel('Maximum Probability')
    plt.ylabel('Count')
    plt.axvline(x=np.mean(max_probs), color='red', linestyle='--', 
                label=f'Mean: {np.mean(max_probs):.3f}')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Analisis Error
    print("\nüîç ANALISIS ERROR:")
    print("-" * 40)
    wrong_predictions = predicted_classes != true_classes
    error_rate = np.mean(wrong_predictions)
    print(f"Error Rate: {error_rate:.4f} ({np.sum(wrong_predictions)} dari {len(predicted_classes)})")
    
    # Analisis confidence pada prediksi salah
    if np.sum(wrong_predictions) > 0:
        wrong_confidences = max_probs[wrong_predictions]
        print(f"Average confidence pada prediksi salah: {np.mean(wrong_confidences):.4f}")
        print(f"Min confidence pada prediksi salah: {np.min(wrong_confidences):.4f}")
    
    # Model Performance Summary
    print(f"\nüìà MODEL PERFORMANCE SUMMARY:")
    print("-" * 40)
    print(f"Overall Accuracy: {report['accuracy']:.4f}")
    print(f"Macro Average F1: {report['macro avg']['f1-score']:.4f}")
    print(f"Weighted Average F1: {report['weighted avg']['f1-score']:.4f}")
    
    # Analisis Overfitting
    train_acc = 0.9977  # Dari output
    val_acc = report['accuracy']
    overfitting_gap = train_acc - val_acc
    
    print(f"\n‚ö†Ô∏è ANALISIS OVERFITTING:")
    print("-" * 40)
    print(f"Training Accuracy: {train_acc:.4f}")
    print(f"Validation Accuracy: {val_acc:.4f}")
    print(f"Overfitting Gap: {overfitting_gap:.4f}")
    
    if overfitting_gap > 0.02:
        print("üö® PERINGATAN: Model mungkin mengalami overfitting!")
        print("Rekomendasi: Tambah dropout, data augmentation, atau kurangi kompleksitas model")
    elif overfitting_gap > 0.01:
        print("‚ö†Ô∏è HATI-HATI: Ada tanda-tanda overfitting ringan")
        print("Rekomendasi: Monitor performa pada data test terpisah")
    else:
        print("‚úÖ Model generalisasi dengan baik")
    
    return predictions, predicted_classes, true_classes

def visualize_sample_predictions(model, val_images, val_labels, class_names, num_samples=8):
    """Visualisasi sampel prediksi"""
    # Ambil sampel random
    indices = random.sample(range(len(val_images)), num_samples)
    sample_images = val_images[indices]
    sample_labels = val_labels[indices]
    
    # Prediksi
    predictions = model.predict(sample_images, verbose=0)
    
    plt.figure(figsize=(16, 8))
    
    for i in range(num_samples):
        plt.subplot(2, 4, i + 1)
        
        # Tampilkan gambar
        plt.imshow(sample_images[i])
        
        # Get true dan predicted labels
        true_class = class_names[np.argmax(sample_labels[i])]
        pred_class = class_names[np.argmax(predictions[i])]
        confidence = np.max(predictions[i])
        
        # Color coding: green untuk benar, red untuk salah
        color = 'green' if true_class == pred_class else 'red'
        
        plt.title(f'True: {true_class}\nPred: {pred_class}\nConf: {confidence:.3f}', 
                 color=color, fontsize=10)
        plt.axis('off')
    
    plt.suptitle('Sample Predictions (Green=Correct, Red=Wrong)', fontsize=14)
    plt.tight_layout()
    plt.show()

def analyze_model_complexity(model):
    """Analisis kompleksitas model"""
    print("\nüèóÔ∏è ANALISIS KOMPLEKSITAS MODEL:")
    print("-" * 40)
    
    total_params = model.count_params()
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    non_trainable_params = total_params - trainable_params
    
    print(f"Total Parameters: {total_params:,}")
    print(f"Trainable Parameters: {trainable_params:,}")
    print(f"Non-trainable Parameters: {non_trainable_params:,}")
    print(f"Model Size: ~{total_params * 4 / (1024**2):.1f} MB")
    
    # Analisis layer
    trainable_layers = sum(1 for layer in model.layers if layer.trainable)
    total_layers = len(model.layers)
    
    print(f"Total Layers: {total_layers}")
    print(f"Trainable Layers: {trainable_layers}")
    print(f"Frozen Layers: {total_layers - trainable_layers}")

# --- PENGATURAN DASAR ---
base_dir = '/kaggle/input/vietnamese-bike-and-motorbike'
IMAGE_SIZE = (160, 160)
BATCH_SIZE = 32
MAX_SAMPLES_PER_CLASS = 3000

if not os.path.exists(base_dir):
    print(f"Error: Direktori tidak ditemukan di '{base_dir}'")
    exit()

print(f"Dataset ditemukan di: {base_dir}")

# --- IDENTIFIKASI FILE RUSAK ---
print("Mengidentifikasi file rusak...")
corrupted_files, valid_files = identify_corrupted_files(base_dir)

if corrupted_files:
    print(f"PERINGATAN: {len(corrupted_files)} file rusak ditemukan!")
    for corrupt_file in corrupted_files[:3]:
        print(f"  - {corrupt_file}")

# --- LOAD DAN PREPARE DATA ---
try:
    # Prepare data
    all_files, all_labels = load_and_prepare_data(valid_files, IMAGE_SIZE, MAX_SAMPLES_PER_CLASS)
    
    # Load images
    all_images, processed_labels = load_images_in_batches(all_files, all_labels, IMAGE_SIZE)
    
    # Convert labels ke categorical
    categorical_labels = to_categorical(processed_labels, num_classes=2)
    
    print(f"Data shape: {all_images.shape}")
    print(f"Labels shape: {categorical_labels.shape}")
    print(f"Class distribution: {np.bincount(processed_labels)}")
    
    # Split data: 70% train, 20% val, 10% test
    train_images, temp_images, train_labels, temp_labels = train_test_split(
        all_images, categorical_labels, 
        test_size=0.3, random_state=42, stratify=processed_labels
    )
    
    val_images, test_images, val_labels, test_labels = train_test_split(
        temp_images, temp_labels, 
        test_size=0.33, random_state=42, 
        stratify=np.argmax(temp_labels, axis=1)
    )
    
    print(f"Training samples: {len(train_images)}")
    print(f"Validation samples: {len(val_images)}")
    print(f"Test samples: {len(test_images)}")
    
    # Buat datasets
    train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
    train_dataset = train_dataset.shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    
    validation_dataset = tf.data.Dataset.from_tensor_slices((val_images, val_labels))
    validation_dataset = validation_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    
    class_names = ['bike', 'motorbike']
    
except Exception as e:
    print(f"Error dalam loading data: {e}")
    exit()

# --- MEMBANGUN MODEL ---
print("\nMembangun model...")

base_model = MobileNetV2(
    input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

# Model architecture dengan regularization yang tepat
inputs = Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3))
x = base_model(inputs, training=False)
x = GlobalAveragePooling2D()(x)
x = Dropout(0.2)(x)
x = Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
x = Dropout(0.3)(x)
outputs = Dense(len(class_names), activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Analisis kompleksitas model
analyze_model_complexity(model)

# --- TRAINING ---
print("\nMemulai training...")

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

callbacks = [
    EarlyStopping(
        monitor='val_loss', 
        patience=5, 
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.5, 
        patience=3, 
        min_lr=1e-7,
        verbose=1
    )
]

try:
    history = model.fit(
        train_dataset,
        epochs=15,
        validation_data=validation_dataset,
        callbacks=callbacks,
        verbose=1
    )
    
    # --- VISUALISASI HASIL TRAINING ---
    plot_training_history(history)
    
    # --- EVALUASI KOMPREHENSIF ---
    predictions, predicted_classes, true_classes = evaluate_model_comprehensive(
        model, val_images, val_labels, class_names
    )
    
    # --- VISUALISASI SAMPLE PREDICTIONS ---
    print("\nüñºÔ∏è SAMPLE PREDICTIONS:")
    visualize_sample_predictions(model, val_images, val_labels, class_names)
    
    # --- EVALUASI PADA TEST SET ---
    if 'test_images' in locals():
        print("\nüß™ EVALUASI PADA TEST SET:")
        print("-" * 40)
        test_loss, test_accuracy = model.evaluate(test_images, test_labels, verbose=0)
        print(f"Test Accuracy: {test_accuracy:.4f}")
        print(f"Test Loss: {test_loss:.4f}")
        
        # Test set confusion matrix
        test_predictions = model.predict(test_images, verbose=0)
        test_pred_classes = np.argmax(test_predictions, axis=1)
        test_true_classes = np.argmax(test_labels, axis=1)
        
        test_cm = confusion_matrix(test_true_classes, test_pred_classes)
        
        plt.figure(figsize=(8, 6))
        sns.heatmap(test_cm, annot=True, fmt='d', cmap='Greens', 
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Test Set Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
    
    # --- FINAL SUMMARY ---
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    final_train_loss = history.history['loss'][-1]
    final_val_loss = history.history['val_loss'][-1]
    
    print(f"\nüéØ RINGKASAN AKHIR:")
    print("="*50)
    print(f"Training Accuracy: {final_train_acc:.4f}")
    print(f"Validation Accuracy: {final_val_acc:.4f}")
    print(f"Training Loss: {final_train_loss:.4f}")
    print(f"Validation Loss: {final_val_loss:.4f}")
    print(f"Overfitting Gap: {final_train_acc - final_val_acc:.4f}")
    
    if 'test_accuracy' in locals():
        print(f"Test Accuracy: {test_accuracy:.4f}")
        print(f"Generalization Gap: {final_val_acc - test_accuracy:.4f}")
    
    print(f"Classes: {class_names}")
    print(f"Total Training Epochs: {len(history.history['accuracy'])}")
    
    # Rekomendasi
    print(f"\nüí° REKOMENDASI:")
    print("-" * 40)
    if final_train_acc - final_val_acc > 0.02:
        print("‚Ä¢ Model overfitting - pertimbangkan lebih banyak regularization")
    elif final_val_acc < 0.95:
        print("‚Ä¢ Accuracy bisa ditingkatkan - coba fine-tuning atau arsitektur lebih kompleks")
    else:
        print("‚Ä¢ Model performance sangat baik!")
    
    if 'test_accuracy' in locals() and abs(final_val_acc - test_accuracy) > 0.02:
        print("‚Ä¢ Ada gap antara validation dan test - cek data leakage")
    
    print("‚Ä¢ Pertimbangkan test pada data real-world untuk validasi lebih lanjut")
    
except Exception as e:
    print(f"Error selama training: {e}")
    print("Coba kurangi batch size atau jumlah sampel.")
```
